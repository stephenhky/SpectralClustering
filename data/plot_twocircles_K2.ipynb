```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
%matplotlib inline
```

```python
points = np.load('/data/hok/testdata/fastsc/two_circles_1000.npy')
```

# FastSC

```python
cluster_file = open('/data/hok/testdata/fastsc/two_circles_1000_output_k2_scoutput.txt', 'r')
cluster_file_reader = csv.reader(cluster_file, delimiter=' ')
cluster_file_reader.__next__()
cluster_results = [(int(rowid), int(cluster)) for rowid, cluster in cluster_file_reader]
cluster_file.close()
```

```python
cluster_results = sorted(cluster_results, key=lambda item: item[0])
cluster_results
```

```python
clusters = np.array([row[1] for row in cluster_results])
clusters
```

```python
plt.scatter(points[:, 0], points[:, 1], c=clusters)
```

# scikit-learn kmeans

```python
from sklearn.cluster import k_means

centroids, labels, intertia = k_means(points, n_clusters=2)
```

```python
plt.scatter(points[:, 0], points[:, 1], c=labels)
```

# scikit-learn SpectralClustering

```python
from sklearn.cluster import SpectralClustering

sklearn_sc = SpectralClustering(n_clusters=2, assign_labels='kmeans', random_state=0).fit(points)

sklearn_sc.labels_
```

```python
plt.scatter(points[:, 0], points[:, 1], c=sklearn_sc.labels_)
```
